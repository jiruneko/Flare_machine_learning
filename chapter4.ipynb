```python
xcol = ['身長', '体重', '年代']
x = df[xcol]
x
```

```python
t = df['派閥']
t
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier(random_state = 0)
model.fit(x, t)
```

```python
taro = [[170, 70, 20]]
taro_df = pd.DataFrame(taro,
                       columns=x.columns)

model.predict(taro_df)
```

```python
matsuda = [172, 65, 20]
asagi = [158, 48, 20]
new_data = pd.DataFrame([matsuda, asagi],
                        columns=x.columns)
model.predict(new_data)
```

```python
model.score(x,t)
```
